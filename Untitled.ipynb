```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
charity_df = pd.read_csv('charity_data.csv')
```

```python
charity_df.head()
```

```python
names = charity_df['NAME']
target = charity_df['IS_SUCCESSFUL']
charity_data = charity_df.drop(columns = ['NAME','IS_SUCCESSFUL'])
```

```python
# Generate our categorical variable list
catagories = charity_data.dtypes[charity_data.dtypes == "object"].index.tolist()
```

```python
# Check the number of unique values in each column
charity_data[catagories].nunique()
```

```python
catagories.remove('CLASSIFICATION')
```

```python
print(catagories)
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(charity_data[catagories]))

# Add the encoded variable names to the DataFrame
encode_df.columns = enc.get_feature_names(catagories)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
charity_data = charity_data.merge(encode_df,left_index=True, right_index=True)
charity_data = charity_data.drop(catagories,1)
charity_data.head()
```

```python
class_counts = charity_data.CLASSIFICATION.value_counts()
class_counts
```

```python
# Determine which values to replace
replace_class = list(class_counts[class_counts < 100].index)

# Replace in DataFrame
for c in replace_class:
    charity_data.CLASSIFICATION = charity_data.CLASSIFICATION.replace(c,"Other")


# Check to make sure binning was successful
charity_data.CLASSIFICATION.value_counts()
```

```python
# Create the OneHotEncoder instance
enc2 = OneHotEncoder(sparse=False)

# Fit the encoder and produce encoded DataFrame
encode_df2 = pd.DataFrame(enc2.fit_transform(charity_data.CLASSIFICATION.values.reshape(-1,1)))

# Rename encoded columns
encode_df2.columns = enc2.get_feature_names(['CLASSIFICATION'])
encode_df2.head()
```

```python
# Merge the two DataFrames together and drop the Country column
charity_data = charity_data.merge(encode_df2,left_index=True,right_index=True).drop("CLASSIFICATION",1)
```
